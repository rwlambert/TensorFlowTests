```python
import tensorflow.contrib.learn as skflow
from sklearn import datasets, metrics
```

```python
def skit_to_df(ds):
    """Simple transform of skitlearn datasets to pandas dataframe"""
    return pd.DataFrame(data=np.c_[ds['data'], ds['target']],
                     columns=ds['feature_names'] + ['target'])
```

```python
def df_to_skit(df, skdict={}):
    """Simple reconstruction of pandas dataframe to new skitlearn dataset
    permits copying metadata"""
    import copy
    skdict = copy.deepcopy(skdict)
    skdict["target"] = df.target.astype(int).values
    skdict["data"] = df[[c for c in df.columns if c!="target"]].values
    skdict["feature_names"]=[c for c in df.columns if c!="target"]
    return skdict

# Decorate for convenience
pd.DataFrame.to_skit = df_to_skit
```

```python
# I can probably make quite a good classifier with simple cuts ...
for i,c in enumerate(['blue','green','red']):
    ii = iris_df[iris_df.target == i]
    plt.scatter(ii["petal width (cm)"] , ii["petal length (cm)"], color=c)

colors = ['blue','green','red']
x = np.linspace(0,2.5,100)
y = x*2.3 + 1
plt.plot(x,y)
import math
def species_guess(width, length):
    """My own guessing function for species, built on the obvious correlation between
    the most obvious separating variables"""
    return np.ceil(np.sqrt(width*width/9+length*length/13.8)*1.4) - 1

plt.plot(x,species_guess(x,y))

iris_df.guess = species_guess(iris_df["petal width (cm)"] , iris_df["petal length (cm)"])

for i,c in enumerate(colors):
    ii = iris_df[iris_df.guess == i]
    plt.scatter(ii["petal width (cm)"] , ii["petal length (cm)"] -2, color=c, marker='x')
```

```python
print sum(iris_df.guess == iris_df.target)/float(len(iris_df))
```

```python
from sklearn import cross_validation
x_train, x_test, y_train, y_test = cross_validation.train_test_split(
    iris_sk.data, iris_sk.target, test_size=0.3, random_state=42)
```

```python
# Infering the real valued columns can take several arguments, including usual dataframes
print skflow.infer_real_valued_columns_from_input(iris_df.T[:-1].T)
print skflow.infer_real_valued_columns_from_input(iris_sk.data)
print skflow.infer_real_valued_columns_from_input(iris_df.T[:-1].T.values)
feature_columns = skflow.infer_real_valued_columns_from_input(x_train)
```

## Data massaging complete, let's train a cool model

```python
classifier = skflow.DNNClassifier(hidden_units=[5, 5, 5], n_classes=3, feature_columns=feature_columns)
```

```python

%timeit classifier.fit(x_train, y_train, steps=50)
```

```python
score = metrics.accuracy_score(y_test, classifier.predict(x_test))
print("Accuracy: %f" % score)
```

```python
iris_df.predict = classifier.predict(iris_df.T[:-1].T.values)
```
